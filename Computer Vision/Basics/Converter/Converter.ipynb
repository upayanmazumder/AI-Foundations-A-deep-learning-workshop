```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files
from IPython.display import Image
```

```python
uploaded = files.upload()
```

```python
Image('image.jpg')
```

```python
import numpy as np
filename = list(uploaded.keys())[0]
image = cv2.imdecode(np.frombuffer(uploaded[filename], np.uint8), cv2.IMREAD_COLOR)
```

```python
print("Size:", image.size)
print("Shape:", image.shape)
print("Data Type:", image.dtype)
```

```python
import matplotlib.pyplot as plt
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(6, 6))
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
_, thresholded_image = cv2.threshold(gray_image, 40, 255, cv2.THRESH_BINARY)
plt.figure(figsize=(6, 6))
plt.imshow(thresholded_image, cmap='gray')
plt.title("Thresholded Image")
plt.axis('off')
plt.show()
```

```python
blurred_image = cv2.GaussianBlur(gray_image, (99, 99), 0)
plt.figure(figsize=(6, 6))
plt.imshow(blurred_image, cmap='gray')
plt.title("Gaussian Blurred Image")
plt.axis('off')
plt.show()
```

```python
median_blurred_image = cv2.medianBlur(gray_image, 5)
plt.figure(figsize=(6, 6))
plt.imshow(median_blurred_image, cmap='gray')
plt.title("Median Blurred Image")
plt.axis('off')
plt.show()
```

```python
edges_canny = cv2.Canny(gray_image, 100, 200)
plt.figure(figsize=(6, 6))
plt.imshow(edges_canny, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis('off')
plt.show()
```
